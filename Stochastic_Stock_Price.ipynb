```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
mu = 0.1  # Tasa de crecimiento esperada
sigma = 0.2  # Volatilidad
S0 = 100  # Precio inicial
dt = 0.01  # Paso de tiempo
T = 1  # Tiempo total de simulación en años
num_simulations = 1  # Número de simulaciones

# Función para simular la ecuación diferencial estocástica
def simulate_stochastic_differential_equation(mu, sigma, S0, dt, T, num_simulations):
    np.random.seed(42)
    timesteps = int(T / dt)
    t = np.linspace(0, T, timesteps + 1)
    S = np.zeros((num_simulations, timesteps + 1))
    S[:, 0] = S0

    for i in range(1, timesteps + 1):
        dW = np.random.normal(0, np.sqrt(dt), num_simulations)
        S[:, i] = S[:, i - 1] * (1 + mu * dt + sigma * dW)

    return t, S

# Simulación
t, simulated_prices = simulate_stochastic_differential_equation(mu, sigma, S0, dt, T, num_simulations)

# Visualización
plt.figure(figsize=(10, 6))
for i in range(num_simulations):
    plt.plot(t, simulated_prices[i, :], label=f'Simulación {i + 1}')

plt.title('Simulación de la Ecuación Diferencial Estocástica')
plt.xlabel('Tiempo (años)')
plt.ylabel('Precio de la Acción')
plt.legend()
plt.grid(True)
plt.show()
```
